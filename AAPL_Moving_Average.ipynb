```python
import datetime
import pandas as pd
import numpy as np
from pandas_datareader import data as web
import fix_yahoo_finance as yf
start = datetime.datetime(2014,1,2)
end = datetime.datetime.today()
##3D Printing
aapl_df = web.DataReader('AAPL','yahoo', start,end)
aapl_df.head()
```

```python
from matplotlib import pylab
from pylab import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#moveing average for close price of AAPL over the past 50 days
aapl_df['MA50_AAPL'] = aapl_df['Adj Close'].rolling(50).mean()
#moveing average for close price of AAPL over the past 200 days
aapl_df['MA200_AAPL'] = aapl_df['Adj Close'].rolling(200).mean()
aapl_df[['Adj Close','MA50_AAPL','MA200_AAPL']].plot()
plt.xlabel('Date')  
plt.ylabel('Close Price($)')  
plt.title('moving average plot')
#depends on whether you are interested in long orshort term investment, are you plan to buy when trend is up or down
```
